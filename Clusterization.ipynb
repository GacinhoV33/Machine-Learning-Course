**DBSCAN algorithm**

**Ex. 1 - comparing DBSCAN, KMeans, Agglomerative**

```python
from sklearn import  datasets
from sklearn.cluster import DBSCAN

# Generating Datasets

n_samples = 1500
noisy_circles = datasets.make_circles(n_samples=n_samples, factor=.5,
                                      noise=.05)
noisy_moons = datasets.make_moons(n_samples=n_samples, noise=.05)
blobs = datasets.make_blobs(n_samples=n_samples, random_state=8)
no_structure = np.random.rand(n_samples, 2), None
 
# Anisotropicly distributed data
random_state = 170
X, y = datasets.make_blobs(n_samples=n_samples, random_state=random_state)
transformation = [[0.6, -0.6], [-0.4, 0.8]]
X_aniso = np.dot(X, transformation)
aniso = (X_aniso, y)
 
# blobs with varied variances
varied = datasets.make_blobs(n_samples=n_samples,
                             cluster_std=[1.0, 2.5, 0.5],
                             random_state=random_state)
```

```python
data_names = ['noisy_circles', 'noisy_moons', 'blobs', 'no_structure', 'aniso', 'varied']
method_names = ['DBSCAN', 'KMeans', 'AgglomerativeClustering']
eps_lst = [0.15, 0.1, 2.75, 0.035, 0.15, 0.55]
min_samples_lst = [20, 20, 10, 5, 15, 12]
clusters_lst = [2, 2, 3, 15, 3, 8]
fig, ax = plt.subplots(6, 3)
fig.set_size_inches([90, 80])
for i, data in enumerate([noisy_circles[0], noisy_moons[0], blobs[0], no_structure[0], aniso[0], varied[0]]):
  # DBSCAN
  clustering = DBSCAN(eps=eps_lst[i], min_samples=min_samples_lst[i])
  y_predict = clustering.fit_predict(data)
  for k in range(max(y_predict)+1):
      ax[i][0].scatter(data[y_predict == k, 0], data[y_predict == k, 1], s=50, c=fake.hex_color(), marker='s', edgecolor='black', label=f'Cluster {k+1}')
      ax[i][0].legend()
      ax[i][0].grid()
      ax[i][0].set_xlabel("Feature 1")
      ax[i][0].set_ylabel("Feature 2")
  ax[i][0].set_title(f"DBSCAN - {data_names[i]}")

  # KMeans 
  kmeans_ = KMeans(n_clusters=clusters_lst[i], init='k-means++', max_iter=100, random_state=0)  
  kmeans_predict = kmeans_.fit_predict(data)
  for k in range(clusters_lst[i]):
      ax[i][1].scatter(data[kmeans_predict == k, 0], data[kmeans_predict == k, 1], s=50, c=fake.hex_color(), marker='s', edgecolor='black', label=f'Cluster {k}')
      ax[i][1].legend()
      ax[i][1].grid()
      ax[i][1].set_xlabel("Feature 1")
      ax[i][1].set_ylabel("Feature 2")

  # Agglomerative Clustering
  model = AgglomerativeClustering(n_clusters=clusters_lst[i], linkage='ward')
  agg_predict = model.fit_predict(data)
  # ax[j][i].set_title(f"Method - {linkage_name}   N of clusters - {clusters_lst[i]}")
  for k in range(clusters_lst[i]):
      ax[i][2].scatter(data[agg_predict == k, 0], data[agg_predict == k, 1], s=50, c=fake.hex_color(), marker='s', edgecolor='black', label=f'Cluster {k}')
      ax[i][2].legend()
      ax[i][2].grid()
      ax[i][2].set_xlabel("Feature 1")
      ax[i][2].set_ylabel("Feature 2")
```
